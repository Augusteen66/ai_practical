```python
# def collision_count(column,row): #row, col
def collision_count(row, column):
    coll = 0
    cr = row
    cc = column
    for j in range(8): # lef - rig
        if j == column: # queen under consideration
            continue
        if board[row][j] == 1 :
            coll += 1
    
    for j in range(8): # top - bot
        if j == row: # queen under consideration
            continue
        if board[j][column] == 1 :
            coll += 1
    
    while(cr < 7 and  cc < 7):
        cc += 1
        cr +=1
        if board[cr][cc] == 1:
            coll += 1
    cr, cc = row, column
    while(cc > 0 and cr > 0):
        cr -= 1
        cc -=1
        if board[cr][cc] == 1:
            coll += 1
    cr, cc = row, column
    while(cc > 0 and cr < 7):
        cr += 1
        cc -=1
        if board[cr][cc] == 1:
            coll += 1
    cr, cc = row, column
    while(cc < 7 and cr > 0):
        cr -= 1
        cc +=1
        if board[cr][cc] == 1:
            coll += 1
            
                     
    return coll 
```

```python
def totalcoll():
    totcoll = 0
    for i in range(8):
        totcoll += collision_count(i,queens[i]) # for each row
    return totcoll
```

### TSP using NN

```python
import copy
inf = float('inf')

class TSP_AI:
    """Traveling Salesman Problem
        -------------------------------------------------------------------
        Traveling Salesman Problem using Nearest Neighbour AI algorithm
    """

    def __init__(self, city_matrix = None, source = 0):
        self.city_matrix = [[0]*4]*4 if city_matrix is None else city_matrix
        self.n : int = len(self.city_matrix)
        self.source : int = source


    def Input(self):
        self.n = int(input('Enter city count : '))

        for i in range(self.n):                                         # Get the distances between cities
            self.city_matrix.append([
                inf if i == j else int(input(f'Cost to travel from city {i+1} to {j+1} : '))
                for j in range(self.n)
            ])

        self.source = int(input('Source: ')) % self.n                   # Get the source city


    def solve(self):
        minCost = inf                                                   # Initially minCost is infinity
        for i in range(self.n):
            print("Path", end='')
            cost = self._solve(copy.deepcopy(city_matrix), i, i)        # Calling solver for each as source city
            print(f" -> {i+1}    :    Cost = {cost}")
            if cost and cost < minCost: minCost = cost                  # If this cost is optimal, save it
        
        return minCost


    def _solve(self, city_matrix, currCity = 0, source = 0):
        if self.n < 2: return 0
        print(f" -> {currCity+1}", end='')

        for i in range(self.n):
            city_matrix[i][currCity] = inf                              # Set all values in the currCity column as infinity (once visited, shouldn't be visited anymore)

        currMin, currMinPos = inf, 0
        for j in range(self.n):
            if currMin > city_matrix[currCity][j]:                      # Get the nearest city to the current city
                currMin, currMinPos = city_matrix[currCity][j], j

        if currMin == inf: return self.city_matrix[currCity][source]    # If currMin is infinity(i.e. all cities have been visited, return cost of moving from this last city to start city to complete the path-loop)
        city_matrix[currCity][currMinPos] = city_matrix[currMinPos][currCity] = inf     # Set distance from currCity to next city and vice versa to infinity
        return currMin + self._solve(city_matrix, currMinPos, source)   # Calling the next recursion for selected city


if __name__ == '__main__':
    city_matrix = [
        [inf, 10,  15,  20],
        [10,  inf, 35,  25],
        [15,  35,  inf, 30],
        [20,  25,  30,  inf]
    ]

    source_city = 0
    tsp = TSP_AI(city_matrix, source_city)
    print(f"Optimal Cost : {tsp.solve()}")
```

#### TSP using A-star

```python
def make_graph(adj_mat, values):
    n = int(input("Enter number of nodes : "))

    for i in range(0, n):

        value = int(input(f"Enter value of node {i + 1} : "))
        values.append(value)
        lst = []
        for i in range(0, n):
            lst.append(0)
        adj_mat.append(lst)
    e = int(input("Enter number of edges : "))
    for j in range(0, e):
        src = int(input("Enter source node : "))
        dest = int(input("Enter destination node : "))
        cost = int(input("Enter cost from src to dest : "))
        adj_mat[src][dest] = cost
        adj_mat[dest][src] = cost
    for k in range(0, n):
        for l in range(k, n):
            if k != l:
                if adj_mat[k][l] == 0:
                    adj_mat[k][l] = 999
                    
def display(n, adj_mat, values):
    for i in range(0, n):
        for j in range(0, n):
            print(adj_mat[i][j])
        print(values[i])
        
        
adj_mat = [[0, 1, 5, 999], [1, 0, 999, 15], [5, 0, 0, 2], [0, 15, 2, 0]]
values = [4, 8, 2, 6]


def a_star(adj_mat, values, start, goal, n):
    openl = set([])
    openl.add(start)
    closel = set([])
    gn = {}
    parent = {}
    for i in range(0, n):
        gn[i] = 0
        parent[i] = -999
    parent[start] = -1
    fn = values[start]
    curr = None
    while (len(openl) > 0):
        fn = 1000
        print(openl)
        print(gn)
        print()
        for e in openl:
            if curr == None or gn[e] + values[e] < fn:
                curr = e
                fn = gn[e] + values[e]
        if curr == goal:
            return fn, parent
        openl.remove(curr)
        succ = []
        for i in range(0, n):
            if curr != i and adj_mat[curr][i] != 999:
                succ.append(i)
        for successor in succ:
            succ_curr_cost = gn[curr] + adj_mat[curr][successor]
            if successor in openl:
                if gn[successor] <= succ_curr_cost:
                    continue
            elif successor in closel:
                if gn[successor] <= succ_curr_cost:
                    continue
                closel.remove(successor)
                openl.add(successor)
            else:
                openl.add(successor)
                parent[successor] = curr
            gn[successor] = succ_curr_cost
        closel.add(curr)
    if (curr != goal):
        return ("Some error")
    
fn, parent = a_star(adj_mat, values, 0, 3, 4)

def print_path(parent, start, goal):
  stack = []
  stack.append(goal)
  curr = parent[goal]
  while curr != -1:
    stack.append(curr)
    curr = parent[curr]
  print("Path is :")
  while len(stack) > 0:
    print(stack.pop())
    
print_path(parent, 0, 3)
```

### TOH

```python
def Hanoi(n,A,B,C):
    count = 1
    if(n==1):
        print("Move ", n, " from ", A, " to ",C)
        return count
    else:
        count += Hanoi(n-1,A,C,B)
        print("Move ", n, " from ", A, " to ",C)
        count += Hanoi(n-1,B,A,C)
        return count
Hanoi(3, "A", "B", "C")
```
